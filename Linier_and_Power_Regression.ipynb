```python
import pandas as pd

# Mengambil data
url= 'https://raw.githubusercontent.com/Rindlin/Regression-Implementation/main/Student_Performance.csv'
data = pd.read_csv(url)

# Memeriksa data
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

# Ekstraksi kolom yang ingin dieksekusi
X_linear = data[['Hours Studied']]
y_linear = data['Performance Index']

# Membuat dan menyesuaikan model regresi linier
linear_model = LinearRegression()
linear_model.fit(X_linear, y_linear)

# Prediksi nilai
y_pred_linear = linear_model.predict(X_linear)

# Memplot hasil
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours Studied', y='Performance Index', data=data, label='Actual data')
plt.plot(data['Hours Studied'], y_pred_linear, color='red', label='Linear regression fit')
plt.xlabel('Hours Studied (TB)')
plt.ylabel('Performance Index (NT)')
plt.title('Linear Regression: Hours Studied vs Performance Index')
plt.legend()
plt.show()

# Keluaran koefisien model
linear_model.coef_, linear_model.intercept_
```

```python
# Mengubah data menggunakan logaritma natural
X_power = np.log(data[['Hours Studied']])
y_power = np.log(data['Performance Index'])

# Membuat dan menyesuaikan model regresi linier pada data yang telah ditransformasi
power_model = LinearRegression()
power_model.fit(X_power, y_power)

# Prediksi nilai dalam skala log
y_pred_power_log = power_model.predict(X_power)

# Transform the predicted values back to the original scale
y_pred_power = np.exp(y_pred_power_log)

# Memplot hasil
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours Studied', y='Performance Index', data=data, label='Actual data')
plt.plot(data['Hours Studied'], y_pred_power, color='green', label='Power regression fit')
plt.xlabel('Hours Studied (TB)')
plt.ylabel('Performance Index (NT)')
plt.title('Power Regression: Hours Studied vs Performance Index')
plt.legend()
plt.show()

# Keluarkan koefisien model (intersep yang dieksponensial)
power_model.coef_, np.exp(power_model.intercept_)
```
